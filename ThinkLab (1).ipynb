```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
data=pd.read_csv("D:\metadata.csv")
data["Re"] = pd.to_numeric(data["Re"], errors="coerce")
data["Rct"] = pd.to_numeric(data["Rct"], errors="coerce")
data = data.dropna(subset=["Re", "Rct"]) 
data["test_id"] = pd.to_numeric(data["test_id"], errors="coerce")
sampled_data = data.sample(1000, random_state=42)
re_values = sampled_data["Re"].values
rct_values = sampled_data["Rct"].values
cycle_counts = sampled_data["test_id"].values
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection="3d")
sc = ax.scatter(re_values, rct_values, cycle_counts, c=cycle_counts, cmap="viridis", s=50)
ax.set_title("3D Impedance Plot (Re(Z), Im(Z), Aging)", fontsize=14)
ax.set_xlabel("Re(Z) (Ohms)", fontsize=12)
ax.set_ylabel("Im(Z) (Ohms)", fontsize=12)
ax.set_zlabel("Cycle Count (Aging)", fontsize=12)
fig.colorbar(sc, ax=ax, label="Cycle Count")
plt.show()
```
